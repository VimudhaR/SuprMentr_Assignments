# Logic builder

```python
# FizzBuzz from 1 to 50
fizz = 0
buzz = 0
fizzbuzz = 0

for i in range(1, 51):
    if i % 3 == 0 and i % 5 == 0:
        print("FizzBuzz", i)
        fizzbuzz += 1
    elif i % 3 == 0:
        print("Fizz", i)
        fizz += 1
    elif i % 5 == 0:
        print("Buzz", i)
        buzz += 1
    else:
        print(i)

print("Fizz count:", fizz)
print("Buzz count:", buzz)
print("FizzBuzz count:", fizzbuzz)
```

# Student Data Manager

```python
students = [
    {"name": "Alice", "marks": 85},
    {"name": "Bob", "marks": 92},
    {"name": "Charlie", "marks": 78},
    {"name": "David", "marks": 88},
    {"name": "Eva", "marks": 95}
]
topper = students[0]
for s in students:
    if s["marks"] > topper["marks"]:
        topper = s
print("Topper:", topper["name"], topper["marks"])
total = 0
for s in students:
    total += s["marks"]
average = total / len(students)
print("Class Average:", average)
for s in students:
    if s["marks"] >= 90:
        grade = "A"
    elif s["marks"] >= 80:
        grade = "B"
    elif s["marks"] >= 70:
        grade = "C"
    else:
        grade = "D"
    print(s["name"], "Grade:", grade)
```

# Data Doctor

```python
names = ["Alice", "Bob", "Charlie", "Bob", "Eva", "alice"]
ages = [23, 24, 22, 24, 25, 23]
scores = [85, None, 78, 24, 95, 85]
cleaned_names = []
cleaned_ages = []
cleaned_scores = []
for i in range(len(names)):
    if names[i].capitalize() not in cleaned_names:
        cleaned_names.append(names[i].capitalize())
        cleaned_ages.append(ages[i])
        cleaned_scores.append(scores[i] if scores[i] is not None else 0) 
print("Cleaned Data:")
for i in range(len(cleaned_names)):
    print(cleaned_names[i], cleaned_ages[i], cleaned_scores[i])
```

# KNN in Real Life (07/03/2026)

```python
users = {
    "Alice": [5, 3, 1],
    "Bob":   [4, 2, 2],
    "Charlie": [1, 5, 5]
}
new_user = [5, 2, 1]
def distance(u1, u2):
    return sum((a-b)**2 for a, b in zip(u1, u2))**0.5
for user in users:
    print(user, "distance:", round(distance(new_user, users[user]), 2))
print("\nRecommendation: Choose movies liked by the closest user (smallest distance).")
```

# Spam Classifier Thinking (10/03/2026)

```python
def is_spam(email):
    spam_keywords = ["free", "win", "offer", "money", "prize"]
    for word in spam_keywords:
        if word in email.lower():
            return True
    if email.count("http") > 2:  # too many links might be spam
        return True
    return False

email_input = input("Enter the email text: ")

if is_spam(email_input):
    print("SPAM email detected!")
else:
    print("This email is NOT spam.")
```

# Customer Segmentation (11/03/2026)

```python
import sklearn
from sklearn.cluster import KMeans
X = [[25, 30], [22, 40], [35, 80], [40, 90], [23, 35], [38, 85]]
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(X)
print("Cluster labels:", kmeans.labels_)
print("Customer Groups:")
print("Cluster 0: Younger customers with low spending")
print("Cluster 1: Older customers with high spending")
```
