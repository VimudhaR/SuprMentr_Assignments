```python
# Shopping Cart using Collections (List of Lists)

shopping_cart = []

# Add items to cart
shopping_cart.append(["Apple", 2])
shopping_cart.append(["Milk", 1])
shopping_cart.append(["Bread", 3])

# Print current items
print("Current items in the shopping cart:")
for item in shopping_cart:
    print("Item:", item[0], "| Quantity:", item[1])

# Print total number of different items
print("Total different items:", len(shopping_cart))
```
